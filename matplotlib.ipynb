```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
yasList = [10,20,30,30,30,40,50,60,70,75]
kiloList = [20,60,80,85,86,87,70,90,95,90]
```

```python
numpyYasList = np.array(yasList)
numpyKiloList = np.array(kiloList)
```

```python
numpyKiloList
```

```python
numpyYasList
```

```python
plt.plot(numpyYasList,numpyKiloList,"b")  ##blue
plt.xlabel("Yas")
plt.ylabel("Kilo")
plt.title("Kilo'nun Yaşa Göre Değişim Grafiği")
#plt.show()
```

```python
numpyDizisi1 = np.linspace(0,10,20)
numpyDizisi1
```

```python
 numpyDizisi2 = numpyDizisi1 **3
numpyDizisi2
```

```python
plt.plot(numpyDizisi1,numpyDizisi2,"g--")
```

```python
plt.plot(numpyDizisi1,numpyDizisi2,"b*-")
```

```python
plt.subplot(1,2,1)
plt.plot(numpyDizisi1,numpyDizisi2,"r*-")
plt.subplot(1,2,2) #1 sıra olacak 2kolon olacak 2.grafik çizimi
plt.plot(numpyDizisi2,numpyDizisi1,"g--")
```

```python
benimFigure = plt.figure()
figureAxes = benimFigure.add_axes([0.4,0.4,0.6,0.6])
figureAxes.plot(numpyDizisi1,numpyDizisi2,"g")
figureAxes.set_xlabel("X Ekseni Veri İsmi")
figureAxes.set_ylabel("Y Ekseni Veri İsmi")
figureAxes.set_title("Grafik Başlığı")
```

```python
figure2 = plt.figure()
eksen1 = figure2.add_axes([0.1,0.1,0.9,0.9])
eksen2 = figure2.add_axes([0.2,0.4,0.3,0.3])
eksen1.plot(numpyDizisi1,numpyDizisi2,"g")
eksen1.set_xlabel("X Ekseni")
eksen1.set_ylabel("Y Ekseni")
eksen1.set_title("Ana Grafik Başlık")

eksen2.plot(numpyDizisi2,numpyDizisi1,"b")
eksen2.set_xlabel("X Ekseni Küçük")
eksen2.set_ylabel("Y Ekseni Küçük")
eksen2.set_title("Küçük Grafik Başlık")
```

```python
(benimFigure,benimEksenler) = plt.subplots(nrows=1,ncols=2)
type(benimEksenler)

for eksen in benimEksenler:
    eksen.plot(numpyDizisi1,numpyDizisi2,"g")
    eksen.set_xlabel("X ekseni")
    
plt.tight_layout()
#benimFigure
```

```python
yeniFigure = plt.figure() #figsize=(6,6) #dpi=100
yeniEksen = yeniFigure.add_axes([0.1,0.1,0.9,0.9])
yeniEksen.plot(numpyDizisi1,numpyDizisi1 ** 2,label="numpyDizisi ** 2 ")
yeniEksen.plot(numpyDizisi1,numpyDizisi1 ** 3,label="numpyDizisi ** 3 ")
yeniEksen.legend(loc=6)
```

```python
yeniFigure.savefig("benimFigure.png",dpi=200)
```

```python
(benimFigure,benimEksen) = plt.subplots()
benimEksen.plot(numpyDizisi1,numpyDizisi2,"g")
benimEksen.plot(numpyDizisi2,numpyDizisi1,color="#C96F23",alpha=0.3)
```

```python
(yeniFigure,yeniEksen) = plt.subplots()
yeniEksen.plot(numpyDizisi1,numpyDizisi1 + 2,color="blue",linewidth=10.0)
yeniEksen.plot(numpyDizisi1,numpyDizisi1 + 3,color="yellow")

yeniEksen.plot(numpyDizisi1,numpyDizisi1 + 4,color="#C96F23",linestyle="--")
yeniEksen.plot(numpyDizisi1,numpyDizisi1 + 5,color="g",linestyle="-.")
yeniEksen.plot(numpyDizisi1,numpyDizisi1 + 6,color="#AB03DF",linestyle=":")

yeniEksen.plot(numpyDizisi1,numpyDizisi1 + 7,color="#000000",linestyle="--",marker="o",markersize=4,markerfacecolor="red")
yeniEksen.plot(numpyDizisi1,numpyDizisi1 + 8,color="#000000",linestyle="--",marker="+",markersize=8)

```
